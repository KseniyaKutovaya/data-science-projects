```python
  import zipfile

# Распаковка train.csv
with zipfile.ZipFile('fashion-mnist_train.csv.zip', 'r') as zip_ref:
    zip_ref.extractall('.')
```

```python
df=pd.read_csv("fashion-mnist_train.csv.zip")
print(df.shape)
```

```python
print(df.shape)
print(df.columns [:5])
```

```python
print(df.dtypes)
```

```python
x=df.drop("label", axis=1)
y=df["label"]
```

```python
x=x/255.0
```

```python
print("Размер x:", x.shape)
print("Размер y:", y.shape)
print("Первые 5 меток:", y[:5].values)
```

```python
x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
print("Обучающая выборка",
x_train.shape,y_train.shape)
print("Тестовая выборка:", x_test.shape,y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(x_train, y_train)
```

```python
model = LogisticRegression(max_iter=1000)
model.fit(x_train, y_train)
```

```python
from sklearn.metrics import accuracy_score

# Предсказываем на тестовых данных
y_pred = model.predict(x_test)

# Считаем точность
accuracy = accuracy_score(y_test, y_pred)
print(f"Точность модели: {accuracy:.3f}")
```
